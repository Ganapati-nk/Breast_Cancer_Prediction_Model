### DATASET : Breast Cancer Wisconsin (Diagnostic) Data Set

### Attribute Information

1) ID number <br>
2) Diagnosis (M = malignant, B = benign)

<h4>Ten real-valued features are computed for each cell nucleus:</h4>

a) radius (mean of distances from center to points on the perimeter) <br>
b) texture (standard deviation of gray-scale values) <br>
c) perimeter <br>
d) area <br>
e) smoothness (local variation in radius lengths) <br>
f) compactness (perimeter^2 / area - 1.0) <br>
g) concavity (severity of concave portions of the contour) <br>
h) concave points (number of concave portions of the contour) <br>
i) symmetry <br>
j) fractal dimension ("coastline approximation" - 1) <br>

The mean, standard error and "worst" or largest (mean of the three
largest values) of these features were computed for each image,
resulting in 30 features. For instance, field 3 is Mean Radius, field
13 is Radius SE, field 23 is Worst Radius.

All feature values are recoded with four significant digits.

Missing attribute values: none

```python
# seperating x and y
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
#spliting train and test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#Scaling Input Columns
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scasled=scaler.transform(x_test)
```

```python
#Encoding Output Columns 
encoder=LabelEncoder()
y_train_encoded=encoder.fit_transform(y_train)
y_test_encoded=encoder.transform(y_test)
```

```python
#Model Creation
lr=LogisticRegression()
lr.fit(x_train_scaled,y_train_encoded)
```

```python
#Prediction
y_pred_test=lr.predict(x_test_scasled)
y_pred_train=lr.predict(x_train_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Accuracy Score For Training Data
accuracy_score(y_pred_train,y_train_encoded)
```

```python
#Accuracy for Test Data
accuracy_score(y_pred_test,y_test_encoded)
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
#Accuracy with KFold Validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)
cv_scores = cross_val_score(lr, x, y, cv=kf, scoring='accuracy')
print(f'Cross-Validation Accuracy: {np.mean(cv_scores):.2f} ± {np.std(cv_scores):.2f}')
```
